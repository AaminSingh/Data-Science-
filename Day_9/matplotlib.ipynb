```python
import matplotlib.pyplot as plt
categories = ['A','B','C','D']
values = [24,46,13,38]

#maximum value find 
max_value = max(values)

#assign the colors
colors = ["green" if value != max_value else "violet" for value in values]

plt.bar(categories,values,color=colors,width=0.8, bottom =0)
plt.title("Bar Chart: Compare categories",fontdict={
    "fontsize":16,
    "fontweight":"bold",
    "color":"red",
    "family":"Arial" #font family
})

plt.xlabel("Category",fontdict={
    "fontsize":16,
    "fontweight":"bold",
    "color":"red",
    "family":"Arial"
})


plt.ylabel("Category",fontdict={
    "fontsize":16,
    "fontweight":"bold",
    "color":"red",
    "family":"Arial"
})
```

```python
domain =['datascience','machine learning','deep learning','artificial intelligence','DSA']
cutoff =[65,55,70,55,85]
plt.plot(domain,cutoff,marker = '*',color= "violet", linestyle = '--',linewidth = 1,markersize = 8,markerfacecolor ="yellow")
plt.title("Line Chart: Compare categories",fontdict={
    "fontsize":16,
    "fontweight":"bold",
    "color":"red",
    "family":"Arial" #font family
})
plt.show()
```

```python
#Line chart- show trends overtime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

#Create a sequence of 30 consecutive dates starting from 2026-01-01
days = pd.date_range("2026-01-01", periods=30)

#Generate 30 random numbers from a normal distribution(mean=0, std =1)
#compute their cumulative sum and shift all values upward by 50
values = np.cumsum(np.random.normal(0,1,30))+ 50

#create line chart
plt.plot(
    days,
    values,
    marker = 'o',
    color = "violet",
    
)
plt.title("Line chart: Trend over time")

plt.xlabel("Date")
plt.ylabel("Value")
#rotate the x-axis labels by 45 degrees for better visibility
plt.xticks(rotation = 30)
plt.show()
```

```python
#Pie Chart- show percentage

from numpy import size


labels = ["Product A","Product B","Product C","Product D"]
sizes = [23,56,34,36]

plt.pie(sizes,labels=labels,autopct="%1.1f%%",startangle=90,colors=["red","green","blue","orange"],shadow=0.1)
plt.title("Pie Chart: Product Sales Distribution")
plt.show()
```

```python
#histgram
data = np.random.normal(loc=60,scale=10,size=1000)

plt.hist(data,bins=10,color="orange",edgecolor="red")
plt.title("Histogram: Distribution of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#Scatter plot-show relationship between two variables 'linewidth=0.8','edges'
study_hours = [1,2,3,4,5,6,7,8]
marks = [35,42,48,55,65,72,80,90]
plt.scatter(study_hours,marks,color="blue",edgecolor="red",linewidth=0.8,marker="*")
plt.title("Study hours vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
#Box plot-Detect outliers and spread
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Group A":np.random.normal(50,5,100),
    "Group B":np.random.normal(60,15,100),
    "Group C":np.random.normal(55,8,100),
})

sns.boxplot(data = df)
plt.title("Box plot:Spread and outliers")
plt.show()
```
