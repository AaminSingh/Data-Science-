```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#first check the list functionality if seaborn 
print(dir(sns))
```

```python
ft = sns.load_dataset('flights')
# Seaborn is a Python library used for data visualization.

# It is built on top of Matplotlib and also provides some built-in datasets for practice.

# Instead of downloading datasets manually, Seaborn lets you load them with one line of code.
```

```python
ft
```

```python
#Bar plot(average passengers per month)
from matplotlib import colors
sns.barplot(data = ft,x='month',y='passengers',palette = 'plasma')
plt.xticks(rotation=45)
plt.title('Average Passengers per Month')
plt.show()
```

```python
#Box plot(average passengers per month)
sns.boxplot(data = ft,x ='month',y='passengers',palette = 'turbo')
plt.xticks(rotation=45)
plt.title('Average Passengers per Month')
plt.show()
```

```python
#violin plot:Month wise passenger

sns.violinplot(data = ft,x='month',y='passengers',palette ='viridis')
plt.xticks(rotation=45)
plt.title('Month wise Passenger')
```

```python
sns.scatterplot(data =ft,x='year',y= 'passengers')
plt.title('Year wise Passenger')
plt.show()
```

```python
#histogram plot
sns.histplot(ft['passengers'],bins=10,kde=True)
plt.title("Passenger Distribution")
plt.show()
```

```python
#kde plot
sns.kdeplot(ft['passengers'],fill=True, palette='viridis')
plt.title("Density plot")
plt.show()
```

```python
#pair plot since the dataset has only one numeric feature(passengers),
#the pair plot is simple
pivot = ft.pivot(index='month',columns = 'year',values = 'passengers')
sns.heatmap(pivot,cmap='YlGnBu',annot=True,fmt='d')
plt.show()
```

```python
#pair plot since the dataset has only one numeric feature(passengers),
#the pair plot is simple
sns.pairplot(ft)
plt.show()
```

```python
sns.countplot(data = ft,x='month',palette = 'plasma')
plt.xticks(rotation=45)
plt.title('Passenger Count per Month')
plt.show()
```

```python
plt.figure(figsize=(12,8))
#1line plot
plt.subplot(2,2,1)
sns.lineplot(data=ft,x='year',y='passengers')
plt.show()
```

```python
sns.lineplot(data=ft,x='year',y='passengers',hue='month')
plt.title('Monthly Passenger Trend')
plt.show()
```
