```python
data = {
    "Name" :["Akil","Ria","Shreya","Sukhi","Sakshi"],
    "age":[23,21,20,19,22],
    "marks":[67,78,96,34,56],
    "hours_study":[3,4,6,1,2]
}

df = pd.DataFrame(data)
print("full data: \n\n",df,sep ="")
```

```python
print("First five rows\n" , df.head(), sep = "")
```

```python
print("Last five rows\n", df.tail(), sep="")
```

```python
print("Dataset information\n", df.info())
```

```python
print("Describe the table\n", df.describe(), sep ="")
```

```python
print("Average Marks \n",df["marks"].mean())
```

```python
print("Maximum marks\n",df["marks"].max())
```

```python
print("Minimum marks\n",df["marks"].min())
```

```python
print("marks more than 85\n",df[df["marks"] >85],sep ="")
```

```python
#correlation between hours study and marks
print("\n correlation")
print(df[["hours_study", "marks"]].corr())
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["hours_study"], df["marks"])
plt.title("Study hours vs Marks")
plt.xlabel("hours studied")
plt.ylabel("marks")
plt.show()
```
