```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
#create dataset
data = {
    "Height":[185,170,168,179,182,188,180,183,180,180,177],
    "Weight":[80,70,65,75,80,85,80,82,78,79,76]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Plot dataset
plt.figure(figsize=(7,6))

plt.scatter(df["Height"], df["Weight"], s=120)

for i in range(len(df)):
    plt.text(df["Height"][i]+0.2,
             df["Weight"][i]+0.2,
             str(i+1))

plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Original Dataset")
plt.grid(True)

plt.show()
```

```python
initial_centroids = np.array([[170, 70], [180, 80]])
```

```python
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1,random_state=42)
kmeans.fit(df)
```
